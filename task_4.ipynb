```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('C:/Users/Nishil Tejani/Downloads/cooler_on_15032024_145432283.json')
df1 = pd.read_json('C:/Users/Nishil Tejani/Downloads/cooler_off_15032024_145723904.json')

data = df["payload"]["values"]
# print(len(data))
data1 = df1["payload"]["values"]
# print(len(data1))

# merged_data = np.concatenate((data, data1), axis=0)
# print(merged_data)
# print(len(merged_data))
```

```python
block_size = 100
step_size = 50
dfs = []
for i in range(0, len(data) - block_size + 1, step_size):
    start_index = i
    end_index = i + block_size
    block_df = data[start_index:end_index]
    print(block_df)
    dfs.append(block_df)
```

```python
block_size = 100
step_size = 50
dfs1 = []
for i in range(0, len(data1) - block_size + 1, step_size):
    start_index = i
    end_index = i + block_size
    block_df1 = data1[start_index:end_index]
    print(block_df1)
    dfs1.append(block_df1)

# print(len(block_df1))
```

```python
res_data = np.array(dfs)
print(res_data)
print(len(res_data))

print(np.shape(res_data))
```

```python
res_data1 = np.array(dfs1)
print(res_data1)
print(len(res_data1))

print(np.shape(res_data1))
```

```python
merged_data = np.concatenate((res_data, res_data1), axis=0)
print(merged_data)
print(len(merged_data))
```

```python
final_dfs = [pd.DataFrame(data_block, columns=['x', 'y', 'z']) for data_block in merged_data]

for i, df_block in enumerate(final_dfs):
    print(f"DataFrame {i}:")
    print(df_block)
# print(len(df_block))
```

```python
min_max_data = []
for df_block in final_dfs:
    min_values = df_block.min()
    max_values = df_block.max()
    min_max_data.append({
        'x_min': min_values['x'],
        'x_max': max_values['x'],
        'y_min': min_values['y'],
        'y_max': max_values['y'],
        'z_min': min_values['z'],
        'z_max': max_values['z']
    })

result_df = pd.DataFrame(min_max_data)
label_df = pd.DataFrame({
    'label': [df["payload"]["label"]] * len(res_data) + [df1["payload"]["label"]] * len(res_data1)
})

num_blocks = len(dfs)
label_df = pd.concat([label_df] * num_blocks, ignore_index=True)

result_df["label"] = label_df
print(result_df)
```
