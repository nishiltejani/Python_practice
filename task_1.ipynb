```python
num_blocks = len(df) // 100
```

```python
for i in range(num_blocks):
    start_index = i * 100
    end_index = start_index + 100
    block_df = df.iloc[start_index:end_index]

    block_file = f"C:/Users/Nishil Tejani/Downloads/block_{i + 1}.csv"
    block_df.to_csv(block_file, index=False, header=False)
```

```python
remaining_data = len(df) % 100
if remaining_data > 0:
    remaining_block_df = df.iloc[-remaining_data:]
    remaining_block_file = f"C:/Users/Nishil Tejani/Downloads/block_{num_blocks + 1}.csv"
    remaining_block_df.to_csv(remaining_block_file, index=False, header=False)

min_values = df.min()
max_values = df.max()

for col in df.columns:
    print(f"Column '{col}': Min={min_values[col]}, Max={max_values[col]}")
```

```python
block_max_values = {}
block_min_values = {}

for i in range(num_blocks):
    block_file = f"C:/Users/Nishil Tejani/Downloads/block_{i + 1}.csv"
    block_df = pd.read_csv(block_file, header=None)

    block_max_values[i] = {}
    block_min_values[i] = {}

    for column in block_df.columns:
        block_max_values[i][column] = block_df[column].max()
        block_min_values[i][column] = block_df[column].min()

for block_num, max_values in block_max_values.items():
    min_values = block_min_values[block_num]
    print(f"Block {block_num + 1}:")
    for column, max_val in max_values.items():
        min_val = min_values[column]
        print(f"  Column '{column}': Min={min_val}, Max={max_val}")
```
