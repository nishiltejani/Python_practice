```python
# Merge

df = pd.DataFrame({"A":[1,2,3,4], "B": [11,12,13,14]})
df1 = pd.DataFrame({"A":[1,2,3,5], "C":[21,21,23,24]})

pd.merge(df,df1,on="A")
```

```python
import pandas as pd

df = pd.DataFrame({"A":[1,2,3,4], "B": [11,12,13,14]})
df1 = pd.DataFrame({"A":[1,2,3,5], "C":[21,21,23,24]})

pd.merge(df,df1, how = "left")   #use "right" also
```

```python
df = pd.DataFrame({"A":[1,2,3,4], "B": [11,12,13,14]})
df1 = pd.DataFrame({"A":[1,2,3,5], "C":[21,21,23,24]})

pd.merge(df,df1, how = "outer", indicator=True) 
```

```python
df = pd.DataFrame({"A":[1,2,3,4], "B": [11,12,13,14]})
df1 = pd.DataFrame({"A":[1,2,3,5], "B":[21,21,23,24]})

pd.merge(df,df1, left_index=True, right_index=True) 
```

```python
df = pd.DataFrame({"A":[1,2,3,4], "B": [11,12,13,14]})
df1 = pd.DataFrame({"A":[1,2,3,5], "B":[21,21,23,24]})

pd.merge(df,df1, left_index=True, right_index=True, suffixes=("name", "id")) 
```
