```python
# Bar Plot

x = ["Python", "Java", "C", "C++", "R"]
y = [100,67,87,89,92]
c=["g", "y", "b", "r", "c"]

plt.title("Marksheet", fontsize=10)
plt.xlabel("Subjects", fontsize=10)
plt.ylabel("Marks", fontsize=10)

plt.bar(x,y, color=c, width=0.6, align="center", edgecolor="m", linewidth=3, linestyle= ":", alpha=0.5, label="Marksheet")
plt.legend()
plt.show()
```

```python
import numpy as np

x = ["Python", "Java", "C", "C++"]
y = [100,67,87,89]
z = [36,48,68,74]

width=0.2
p = np.arange(len(x))
p1= [j+width for j in p]

plt.title("Marksheet", fontsize=10)
plt.ylabel("Subjects", fontsize=10)
plt.xlabel("Marks", fontsize=10)

plt.barh(p,y, width, color="y", label="Student1")
plt.barh(p1,z, width, color="r", label="Student2")

plt.yticks(p+width/2, x)  # we can also add "rotation" on x named list
plt.legend()
plt.show()
```

```python
# Scatter Plot

x = [3,2,4,5,1,2,7,6]
y = [5,2,7,1,3,6,4,3]
c = ["r","y","g","b","c","r","g","b"]
sizes=[500,150,300,50,100,200,210,400]
plt.scatter(x,y, s=sizes, color=c, alpha=0.5, marker="^")  #we can also add edgecolor to marker parameter

plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
x = [3,2,4,5,1,2,7,6]
y = [5,2,7,1,3,6,4,3]
y1 = [3,2,1,5,4,3,6,7]

colors = [10,30,35,24,45,50,20,40]
sizes=[500,150,300,50,100,200,210,400]

plt.scatter(x,y, s=sizes, c=colors, alpha=0.5, cmap="Accent_r")
plt.scatter(x,y1, s=sizes, color="b")

plt.colorbar()
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
# pie plot

x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex = [0.4,0.0,0.0,0.0]

plt.title("Pie Plot")
plt.pie(x, labels=y, explode=ex, autopct="%0.2f%%", shadow=True, radius=1, labeldistance=1.1, startangle=0
        , counterclock=False, wedgeprops={"linewidth":5})
plt.legend(loc=2)
plt.show()
```

```python
# Stem plot

x = [1,2,3,4,5,6]
y = [2,3,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="yo", basefmt="g", label="python")
plt.legend()
plt.show()
```

```python
# Box Plot

x = [10,20,30,40,50,60,70, 120]
# 120 is shows as an outlier , if we want to joint outlier we need to add "whis=2" so it is join to that outlier
plt.boxplot(x, vert=True, widths=0.2, labels=["Python"]
            , showmeans=True, sym="g+", boxprops=dict(color="r"), capprops=dict(color="r"), whiskerprops=dict(color="r")
            , flierprops=dict(markeredgecolor="y"))  # we add notch=True for notch plot instead of box plot
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
x1 = [10,20,30,40,50,60,90]

y=[x,x1]
plt.boxplot(y, vert=True, widths=0.2, labels=["Python", "C"]
            , showmeans=True, sym="g+", boxprops=dict(color="r"), capprops=dict(color="r"), whiskerprops=dict(color="r")
            , flierprops=dict(markeredgecolor="y"))

plt.show()
```

```python
# Stack plot

x=[1,2,3,4,5]
area=[2,3,4,1,5]
area1=[5,2,2,3,5]
area2=[3,4,1,5,2]

l=["area", "area1", "area2"]
plt.stackplot(x, area, area1, area2, labels=l, colors=["g","m","c"], baseline="zero")
plt.legend(loc=2)
plt.show()
```

```python
# Step plot

x=[1,2,3,4,5]
y=[2,3,4,5,4]

plt.step(x,y,color="c", marker="o", ms=10, mfc="g", label="python")
plt.grid()
plt.legend(loc=2)
plt.show()
```

```python
# Fill_Between

x=[1,2,3,4,5]
y=[1,3,2,5,4]

plt.plot(x,y, color="r")
plt.fill_between(x,y)
plt.show()
```
