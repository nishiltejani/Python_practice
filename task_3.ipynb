```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("C:/Users/Nishil Tejani/Downloads/cooler_on_15032024_145432283.json")
data = df["payload"]["values"]
print(data)
```

```python
block_size = 100
step_size = 50
dfs = []
for i in range(0, len(data) - block_size + 1, step_size):
    start_index = i
    end_index = i + block_size
    block_df = data[start_index:end_index]
    print(block_df)
    dfs.append(block_df)
```

```python
data1 = np.array(dfs)
print(data1)
print(len(data1))
```

```python
dfs = [pd.DataFrame(data_block, columns=['x', 'y', 'z']) for data_block in data1]

for i, df_block in enumerate(dfs):
    print(f"DataFrame {i}:")
    print(df_block)
```

```python
min_max_data = []
for df_block in dfs:
    min_values = df_block.min()
    max_values = df_block.max()
    min_max_data.append({
        'x_min': min_values['x'],
        'x_max': max_values['x'],
        'y_min': min_values['y'],
        'y_max': max_values['y'],
        'z_min': min_values['z'],
        'z_max': max_values['z']
    })

result_df = pd.DataFrame(min_max_data)

print(result_df)
```
