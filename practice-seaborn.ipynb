```python
stats1 = pd.read_excel("Messi and CR7.xlsx", sheet_name='Messi club')
stats2 = pd.read_excel("Messi and CR7.xlsx", sheet_name='Messi national')
stats3 = pd.read_excel("Messi and CR7.xlsx", sheet_name='CR7 club')
stats4 = pd.read_excel("Messi and CR7.xlsx", sheet_name='CR7 national')
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.scatterplot(x='Total Apps', y='Total Goals', hue='Division', data=stats1, 
        style='Club', markers={'Barcelona':'o','PSG':'*'}, 
        size='Club', sizes=(200,50), palette='gist_ncar', alpha=1)
plt.title('Club goals by Messi')

plt.subplot(1,2,2)
sns.scatterplot(x='Total Apps', y='Total Goals', hue='Division', data=stats3, 
        style='Club', size='Club', sizes=(200,50), 
        palette='gist_ncar', alpha=1)
plt.title('Club goals by Ronaldo')

plt.show()
```

```python
# Heatmap
import numpy as np
x1 = np.linspace(1,10,10).reshape(2,5)
display('x1', x1)
anar = np.array([['a 00','a 01','a 02','a 03','a 04'],
                 ['a 10','a 11','a 12','a 13','a 14']])

sns.heatmap(x1, annot=anar, fmt='s', linewidths=5, linecolor='b')
plt.show()
```

```python
# Countplot

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(x='Division', hue='Club', data=stats1, 
            palette='winter_r', saturation=0.9, alpha=0.7)
plt.xticks(rotation=30)
plt.title('no. of seasons played by Messi')

plt.subplot(1,2,2)
sns.countplot(x='Division', hue='Club', data=stats3, 
            palette='winter_r', saturation=0.9, alpha=0.7)
plt.xticks(rotation=30)
plt.title('no. of seasons played by Ronaldo')
plt.show()
```

```python
#violinplot

plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
sns.violinplot(x='Division', y='Total Goals', data=stats1, hue='Club', 
        linewidth=2, palette='Set1', order=['La Liga','Ligue 1'], 
        saturation=0.6, split=False, scale='count', inner='box')
plt.xticks(rotation=30)
plt.title('Total club goals by Messi')

plt.subplot(1,2,2)
sns.violinplot(x='Division', y='Total Goals', data=stats3, hue='Club', 
        linewidth=2, palette='Set1', saturation=0.6, 
        split=False, scale='count', inner='box')
plt.xticks(rotation=30)
plt.title('Total club goals by Ronaldo')
plt.show()
```

```python
# Pairplot

display(stats1.head())

sns.pairplot(stats1, vars=['League Apps','Total Apps','League Goals','Total Goals'], 
            hue='Club', hue_order=['PSG','Barcelona'], 
            palette='rocket_r', diag_kind='hist', markers=['<','o'])
plt.show()
```

```python
# stripplot

plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
sns.stripplot(x='Division', y='Total Goals', data=stats1, hue='Club', jitter=1, 
            palette='rocket', linewidth=0.5, edgecolor='blue', size=5, alpha=0.9)
plt.xticks(rotation=30)
plt.title('Total club goals by Messi')

plt.subplot(1,2,2)
sns.stripplot(x='Division', y='Total Goals', data=stats3, hue='Club', jitter=1, 
            palette='rocket', linewidth=0.5, edgecolor='blue', size=5, alpha=0.9)
plt.xticks(rotation=30)
plt.title('Total club goals by Ronaldo')
plt.show()
```

```python
# boxplot

plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
sns.boxplot(x='Division', y='League Goals', data=stats1, hue='Club', order=['La Liga','Ligue 1'], 
            showmeans=True, meanprops={'marker':'+','markeredgecolor':'g'}, linewidth=2, orient='v')
plt.xticks(rotation=30)
plt.title('League goals by Messi')

plt.subplot(1,2,2)
sns.boxplot(x='Division', y='League Goals', data=stats3, hue='Club', 
            showmeans=True, meanprops={'marker':'+','markeredgecolor':'g'}, linewidth=2, orient='v')
plt.xticks(rotation=30)
plt.title('League goals by Ronaldo')

plt.show()
```
